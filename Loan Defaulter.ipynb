```python
null_counts = df.isnull().sum()
print(null_counts)
```

```python

null_counts_percentage = df.isnull().mean() *100
print(null_counts_percentage)
```

```python
null_counts_percentage.sort_values(ascending=False)
```

```python
high_null_values = null_counts_percentage[null_counts_percentage.values >= 47].index.tolist()
```

```python
len(high_null_values)
```

```python
df.drop(high_null_values, axis=1, inplace= True)
```
